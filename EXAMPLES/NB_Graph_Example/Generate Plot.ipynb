```python
import re
import matplotlib.pyplot as plt
from matplotlib.transforms import offset_copy as oc
%matplotlib inline 

# Subset of matplotlib rcParams that should be different for the inline backend.
# c.InlineBackend.rc = {'font.size': 10, 'figure.figsize': (6.0, 4.0), 'figure.facecolor': 'white', 'savefig.dpi': 72, 'figure.subplot.bottom': 0.125, 'figure.edgecolor': 'white'}
#pylab.rcParams['figure.figsize'] = (10.0, 8.0)

def drawIt(txt, x_val, title, x_scale=None, y_scale0=None, y_scale1=None):
    throughPut = []
    responseTime = []
    threads = []
    sizes = []
    
    for line in txt.split('\n'): 
        line = line.strip()
        line = line.replace(" ", "")   
        if "RUN_FIXED" in line or "RUN_MIXED" in line:
            line = re.findall(r"[\w]+", line)  
            threads.append(int(line[2]))
            #sizes.append(round(float(int(line[4]))/1024, 4))
            sizes.append(int(line[4]))
        if "Throughput" in line:            
            i = line.index('=', 0, len(line))
            j = line.index('bytes/sec', 0, len(line))
            throughPut.append(round(float(line[i+1:j])/1048576, 2))
        if "Avg.ResponseTime" in line:            
            i = line.index('=', 0, len(line))
            j = line.index('secs', 0, len(line))            
            responseTime.append(round(float(line[i+1:j]), 3))
        
    fig, ax0 = plt.subplots(figsize=(4, 4))

    if x_val == 'threads':
        x_val_data = threads
        ax0.set_xlabel("Threads")
    elif x_val == 'sizes':
        x_val_data = sizes
        ax0.set_xlabel("File Size (bytes)")
        
    ax0.grid(True)
    plt.title(title, size=12)
        
    ax0.plot(x_val_data, throughPut, color="green")
    ax0.set_ylabel("Throughtput (MiB/sec)", color="green")
    ax1 = ax0.twinx()
    ax1.plot(x_val_data, responseTime, color="red")
    ax1.set_ylabel("Response Time (sec)", color="red")

    if x_scale:
        ax0.set_xscale('log', basex=x_scale)
        ax1.set_xscale('log', basex=x_scale)
    if y_scale0:
        ax0.set_yscale('log', basey=y_scale0)
    if y_scale1:
        ax1.set_yscale('log', basey=y_scale1)
        
    trans_offset0 = oc(ax0.transData, fig=fig, x=0.05, y=-0.10, units='inches')
    for x, y in zip(x_val_data, throughPut):
        ax0.plot((x,), (y,), 'ro', color="green")
        ax0.text(x, y, '%d, %.2f' % (x, y), transform=trans_offset0, color='green')
    
    trans_offset1 = oc(ax1.transData, fig=fig, x=0.05, y=-0.10, units='inches')
    for x, y in zip(x_val_data, responseTime):
        ax1.plot((x,), (y,), 'ro', color="red")
        ax1.text(x, y, '%d, %.3f' % (x, y), transform=trans_offset1, color='red')
        
    #plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1.0)
    plt.show()
    col_width = 12
    header = (x_val, 'Tpt (MiB/s)', 'RT (s)')
    print "".join(str(word).ljust(col_width) for word in header)
    for row in zip(x_val_data, throughPut, responseTime):
        print "".join(str(word).ljust(col_width) for word in row)
```
